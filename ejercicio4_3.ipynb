```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importar dataset titanic desde seaborn
titanic = sns.load_dataset("titanic")
titanic
```

# 1.Confeccione un gráfico de barras para visualizar hombres y mujeres que han sobrevivido.

```python
mascara_sobrevivientes = titanic['survived'] == 1
titanic_sobrevivientes = titanic[mascara_sobrevivientes]

ax = sns.countplot(
    x="sex",
    data=titanic_sobrevivientes,
    #hue= 'survived',
    palette="colorblind"
)

# agregar titulo
plt.title("Distribución de pasajeros sobrevivientes por sexo")

# agregar etiqueta de valores
plt.xlabel("Sexo")
plt.ylabel("Cantidad de pasajeros")
plt.show()

```

# 2. Cambiar el color del gráfico.

```python
# cambiar color del grafico anterior
custom_palette = ["#FF0000", "#00FF00", "#0000FF", "#FFFF00"]
ax = sns.countplot(
    x="sex",
    data=titanic_sobrevivientes,
    # hue= 'survived',
    palette=['red','blue']
)

# agregar titulo
plt.title("Distribución de pasajeros sobrevivientes por sexo")

# agregar etiqueta de valores
plt.xlabel("Sexo")
plt.ylabel("Cantidad de pasajeros")
plt.show()
```

# 3. Gráfico barras agrupados

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

sns.countplot(
    x="sex",
    data=titanic_sobrevivientes,
    hue= 'pclass',
    palette="colorblind", ax = axs[0]
)
# agregar titulo
axs[0].set_title('Distribución de pasajeros sobrevivientes agrupados por sex y pclass')

sns.countplot(
    x="sex",
    data=titanic_sobrevivientes,
    hue= 'embarked',
    palette="colorblind", ax = axs[1]
)
# agregar titulo
axs[1].set_title('Distribución de pasajeros sobrevivientes agrupados por sex y embarked')
```
