```python
from scipy.stats import norm
import numpy as np
```

$$n = \frac{Z^2\cdot p\cdot(1-p)}{e^2}$$

```python
# n = ??
alfa = 0.05
e = 0.02
p = 0.5
Z = norm.ppf(1-alfa/2)
n = (Z**2*p*(1-p)/e**2)

print(f"El valor minimo de n es: {int(np.ceil(n))}")
```

$$ \bar{x} \pm Z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}} $$

```python
u = 100
s = 12
alfa = 0.02
l_inf = 90
l_sup = 100

# P(90<=X<=100) = 0.98

Z = norm.ppf(1-alfa/2)

n = ((Z * s)/(l_sup-l_inf))**2

print(f"El valor minimo de n es: {int(np.ceil(n))}")
```

```python
s = 3.2
n = 64
u = 32.5
l_inf = 31.5
l_sup = 33.5

# P( (31.5/(3.2/sqrt(64)) * Z < = X <= (33.5/(3.2/sqrt(64)) * Z) = 1-alfa
# 2*Z * (33.5-32.5)/(6/sqrt(64)) = 1-alfa

Z = (l_sup-l_inf)/(s/np.sqrt(n)) /2
P_Z = norm.cdf(Z) # estadistico de prueba Z

# P(Z<=2.5) = 1-alfa/2  # se divide en dos por ser una distribucion de dos colas

alfa = (1-P_Z)*2 # por ser de dos colas

nivel_confianza = 1-alfa

print(f"El nivel de confianza necesario para obtener una media poblacional entre 33.5 y 31.5 es: {nivel_confianza}")
```
