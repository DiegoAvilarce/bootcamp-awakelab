```python
import pandas as pd

# crear la data para el ejercicio
data = {'X': [189, 190, 208, 227, 239, 252, 257, 274, 293, 308, 316],
        'Y': [402, 404, 412, 425, 429, 436, 440, 447, 458, 469, 469]}

df = pd.DataFrame(data)
print(df)
```

# 1)

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Ajustar el modelo de regresión lineal
modelo_rl = LinearRegression()
x = df.X.values # para obtener un arreglo del vector unidemnsional o serie
y = df.Y.values
modelo_rl.fit(x.reshape(-1,1),y) # el -1 es para tomar todos los registros del vector y el 1 es porque es solo una variable
```

```python
# Graficar la dispersión de puntos de datos
plt.scatter(x, y,label="Datos")
plt.xlabel("x")
plt.ylabel("y")

# Calcular las predicciones de la recta de regresión
y_pred = modelo_rl.predict(x.reshape(-1, 1))
# Graficar la recta de regresión
plt.plot(x, y_pred, color="red",label="Recta Ajuste")
plt.legend()
plt.show()
```

```python
# coeficientes del modelo
pendiente = modelo_rl.intercept_
intercepto =  modelo_rl.coef_[0]

print(f"La ecuación de la recta seria: \nY = {round(intercepto,4)} + {round(pendiente,4)}X")
```

```python
# generar matriz de correlacion de Person
corr_matrix = df.corr(method='pearson')
corr_matrix
```

```python
corr_coef = df['X'].corr(df['Y'])
corr_coef
print(f"El coeficiente de correlación de Pearson entre X e Y es: {round(corr_coef,4)}")

# Del coeficiente de correlación se infiere que las variables x e y están altamente correlacionadas de forma positiva
# es decir que a medida que aumenta x, tambien aumenta y en proporciones casi similares
```
