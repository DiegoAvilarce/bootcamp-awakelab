```python
import pandas as pd
data = {'Metros': [1, 2, 3, 4, 5, 6, 7, 8],
        'Ninos': [2, 6, 10, 5, 10, 3, 2, 2]}

df = pd.DataFrame(data)

# esta matriz intenta corregir el error de la matriz anterior
valores = []
for i in range(len(df.Metros)):
    for j in range(df.Ninos[i]):
        valores.append(df.Metros[i])
df_valores = pd.DataFrame(valores, columns=['Metros'])
```

# A)

```python
# Tabla de frecuencias Ninos
frecuencia = df.Ninos.value_counts()
norm_frecuencia = df.Ninos.value_counts(normalize=True)

# calcular frecuencia acumulada
frecuencia_acumulada = frecuencia.cumsum()
norm_frecuencia_acumulada = norm_frecuencia.cumsum()
```

```python
frecuencia.plot(kind='bar',title='Frecuencia Absoluta Ninos')
```

```python
norm_frecuencia.plot(kind='bar',title='Frecuencia Relativa Ninos')
```

```python
frecuencia_acumulada.plot(kind='bar',title='Frecuencia Absoluta Acumulada Ninos')
```

```python
norm_frecuencia_acumulada.plot(kind='bar',title='Frecuencia Relativa Acumulada Ninos')
```

# B)

```python
# calcular media, mediana y moda
media = df.Ninos.mean()
mediana = df.Ninos.median()
moda = df.Ninos.mode()
cuartiles = df.Ninos.quantile([0,0.25,0.5,0.75,1])
media,mediana,moda,cuartiles
```

```python
# calcular desviacion estandar y varianza
desvest = df.Ninos.std()
varianza = df.Ninos.var()

desvest, varianza
```
