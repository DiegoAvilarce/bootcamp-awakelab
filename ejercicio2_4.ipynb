```python
from scipy.stats import norm

u = 10   # media
o = 2    # desviación estándar
lower = u - 3 * o
upper = u + 3 * o

P = norm.cdf(upper, u, o) - norm.cdf(lower, u, o)

print("La probabilidad es:", P)
```
