Aprendizaje supervisado:
Son algoritmos que intentan predecir un resultado a partir de un conjunto de datos cuya respuesta ya es conocida. Es decir, que son modelos que pretenden predecir una variable explicada a través de variables explicativas.

Aprendizaje sins supervisión:
Son algoritmos que intentan encontrar patrones en los datos proporcionados, sin tener una variable de respuesta a predecir. En otras palabras,
lo que busca este tipo de algoritmos es analizar y encontrar agrupaciones entre los mismos datos de forma que permita etiquetarlos en categorías o clusters. 

Aprendizaje por refuerzo:
Son algoritmos que se utilizan cuando hay múltiples variables que considerar ante escenarios o toma de decisiones complejas. Son útiles
cuando hay muchos factores que llevan a tomar una decisión, por lo que el aprendizaje en estos casos se basa en una asignación de recompensas o castigos sobre las distintas decisiones que se pueden tomar, buscando por lo tanto la maximización de las recompensas a través del ensayo y error. 

Algoritmos de regresión:
Son algoritmos que buscan predecir el resultado de una variable númerica explicada mediante una relación matemática que exista entre sus variables explicativas. Por lo que se puede incluir este tipo de algritmos dentro del aprendizaje supervisado, ya que se debe conocer previamente el valor real de la variable explicada para poder crear un modelo matemático que se ajuste a dichos resultados. 

Algoritmos de agrupación:
Son algortitmos utilizados para encontrar patrones y similitudes entre los datos con tal de permitir agruparlos en clusters. En donde los datos dentro de un cluster sean lo mas homogeneos posibles y los datos entre distintos clusters sean lo mas diferentes, con respecto a caracteristicas especificas. Estos algoritmos se incluyen dentro del aprendizaje no supervisado. 

Algoritmos de Aprendizaje Profundo:
Son algoritmos que buscan modelar predicciones complejas mediante transformaciones iterativas y no lineales. Se basan en lo que se conoce como redes neuronales, en donde los datos viajan entre multiples neuronas y capas que los procesan para generar respuestas complejas e identificar patrones complejos en los datos. Estos datos pueden se procesadas por multiples capas o neuronas, permitiendo así predecir resultados mas abstractos o complejos.
