```python
import pandas as pd
```

```python
# Lee csv
datos = pd.read_csv('rrhh.csv')
```

```python
datos.info()
```

```python
# Castear Estado y Sexo como factor
datos['Estado'] = datos['Estado'].astype('category')
datos['Sexo'] = datos['Sexo'].astype('category')
datos['Estado.Civil'] = datos['Estado.Civil'].astype('category')

```

```python
datos.describe()
```

```python
# Genera lista de valores nulos
datos.isnull().sum()
```

```python
# Generar lista de valores nan
datos.isna().sum()
```

```python
# Del comando describe se obtiene que no hay valores negativos en las variables num√©ricas
# Tampoco hay valores nulos o nan.
```

```python
# Generar bocplot con datos atipicos de todas las variables num√©ricas
import matplotlib.pyplot as plt
import seaborn as sns

fig, axs = plt.subplots(ncols=3, nrows=2, figsize=(15, 10))
sns.boxplot(x=datos['Edad'], ax=axs[0, 0])
sns.boxplot(x=datos['Ratio.Pago'], ax=axs[0, 1])
sns.boxplot(x=datos['Salario'], ax=axs[0, 2])
sns.boxplot(x=datos['Dias.trabajados'], ax=axs[1, 0])
sns.boxplot(x=datos['Ausencias'], ax=axs[1, 1])
```

```python
# Por criterio de rangos intercuartiles de los boxplots, la variable Salario tiene mas valores at√≠picos que el resto.
# Las variables Ausencias y Ratio.pago no tienen valores at√≠picos.
```

```python
# Graficar pairs plot de las variables num√©ricas con pearson como coeficiente de correlaci√≥n y ajuste lineal
sns.pairplot(datos, x_vars=['Edad', 'Ratio.Pago', 'Salario','Dias.trabajados','Ausencias'], y_vars='Estado', 
             kind='reg', diag_kind='kde',height=5, aspect=0.7)

```
