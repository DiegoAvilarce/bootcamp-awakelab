```python
# Generar bocplot con datos atipicos de todas las variables numéricas
import matplotlib.pyplot as plt
import seaborn as sns

fig, axs = plt.subplots(ncols=3, nrows=2, figsize=(15, 10))
sns.boxplot(x=datos['Edad'], ax=axs[0, 0])
sns.boxplot(x=datos['Ratio.Pago'], ax=axs[0, 1])
sns.boxplot(x=datos['Salario'], ax=axs[0, 2])
sns.boxplot(x=datos['Dias.trabajados'], ax=axs[1, 0])
sns.boxplot(x=datos['Ausencias'], ax=axs[1, 1])
```

```python
# Promedio, desvest y coef de variación de las variables numéricas

media_edad = datos['Edad'].mean()
media_ratio = datos['Ratio.Pago'].mean()
media_salario = datos['Salario'].mean()
media_dias = datos['Dias.trabajados'].mean()
media_ausencias = datos['Ausencias'].mean()

desvest_edad = datos['Edad'].std()
desvest_ratio = datos['Ratio.Pago'].std()
desvest_salario = datos['Salario'].std()
desvest_dias = datos['Dias.trabajados'].std()
desvest_ausencias = datos['Ausencias'].std()

coef_var_edad = desvest_edad/media_edad
coef_var_ratio = desvest_ratio/media_ratio
coef_var_salario = desvest_salario/media_salario
coef_var_dias = desvest_dias/media_dias
coef_var_ausencias = desvest_ausencias/media_ausencias

print(f"La edad promedio es {media_edad:.2f}, su desviación estándar es {desvest_edad:.2f} y su coeficiente de variación es {coef_var_edad:.2f}")
print()
print(f"El ratio de pago promedio es {media_ratio:.2f}, su desviación estándar es {desvest_ratio:.2f} y su coeficiente de variación es {coef_var_ratio:.2f}")
print()
print(f"El salario promedio es {media_salario:.2f}, su desviación estándar es {desvest_salario:.2f} y su coeficiente de variación es {coef_var_salario:.2f}")
print()
print(f"Los días trabajados promedio son {media_dias:.2f}, su desviación estándar es {desvest_dias:.2f} y su coeficiente de variación es {coef_var_dias:.2f}")
print()
print(f"Las ausencias promedio son {media_ausencias:.2f}, su desviación estándar es {desvest_ausencias:.2f} y su coeficiente de variación es {coef_var_ausencias:.2f}")
```

```python
# En primera instancia al analizar el graficos Pairs, se puede inferir que las variables Edad, Dias.trabajados y Ratio.Pago podrian ser 
# mas signifactivas para predecir el estado de un empleado. Ya que presentan pendientes mas pronunciadas y una mayor correlación con el estado. 
# Por ejemplo, a mayor edad es mas probable que el empleado tenga estado 1 (desvinculado). En cambio, a mayor dias de trabajo y a mayor
# Ratio de pagos la posibilidad de que el empleado este desvinculado disminuye.
# Por otro lado, del mismo gráfico se puede apreciar que el Salario y las Ausencias no son tan significativas para predecir el estado de 
# un empleado, ya que presentan una pendiente casi paralela al eje x, es decir que tienen una correlación muy baja con la variable explicada.

# Del boxplot de Dias trabajados se puede observar claramente que la caja para los valores de estado 1 está mas a la izquierda que la caja 
# del estado 0, esto quiere decir que los empleados desvinculados trabajaron menos dias que los empleados vinculados. Por lo que se puede
# inferir que las desvinculaciones se producen en aquellos empleados que trabajaron menos dias o que bien tienen menos antiguedad en la empresa.

# Por último cabe destacar que los dias trabajados tiene el mayor coeficiente de variación. Por lo que se puede inferir que 
# los empleados del set de datos trabajan una cantidad de dias muy variada. Caso contrario ocurre con la edad, la cual tiene el menor
# coeficiente de variación, por lo que se puede decir que los empleados del set de datos en promedio no tienen mucha diferencia 
# de edad entre ellos.
```

```python
for valor in datos['Sexo'].unique():
    total = datos['Sexo'].shape[0]
    estado1 = datos[datos['Sexo'] == valor]['Estado'].count()
    print(f"La probabilidad de estar desvinculado dado sexo {valor} es {estado1/total:.2f}")
```

```python
for valor in datos['Estado.Civil'].unique():
    total = datos['Estado.Civil'].shape[0]
    estado1 = datos[datos['Estado.Civil'] == valor]['Estado'].count()
    print()
    print(f"La probabilidad de estar desvinculado dado estado civil {valor} es {estado1/total:.2f}")
```

```python
for valor in datos['Departamento'].unique():
    total = datos['Departamento'].shape[0]
    estado1 = datos[datos['Departamento'] == valor]['Estado'].count()
    print()
    print(f"La probabilidad de estar desvinculado dado departamento {valor} es {estado1/total:.2f}")
```

```python
for valor in datos['Desempeño'].unique():
    total = datos['Desempeño'].shape[0]
    estado1 = datos[datos['Desempeño'] == valor]['Estado'].count()
    print()
    print(f"La probabilidad de estar desvinculado dado desempeño {valor} es {estado1/total:.2f}")
```

```python
diccionario = {}
for valor in datos['Posicion'].unique():
    total = datos['Posicion'].shape[0]
    estado1 = datos[datos['Posicion'] == valor]['Estado'].count()
    diccionario[valor] = round(estado1/total,2)

# transformar diccionario a dataframe
df = pd.DataFrame(diccionario.items(), columns=['Posicion', 'Probabilidad'])
df = df.sort_values(by=['Probabilidad'], ascending=False)
df
```

```python
# Si miramos los gráficos podemos observar muy a simple vista que la mayoría de las desvinculaciones están en el departamento 
# de producción con un 67% del total de desvinculaciones. Por otro lado, el departamento Executive Office es el que presenta menos 
# desvinculaciones.

# También se puede ver que las posiciones con mas despidos son Production Technician I y Production Technician II, con un 44% y 18% 
# respectivamente. Esto se condice con la observaciones anterior, donde se decía que el departamento de producción es el que más despidos
# tiene. 

# El sexo parece no ser una variable significativa para predecir el estado de un empleado, ya que la probabilidad de estar desvinculado
# dado el sexo no es lo suficientemente diferente entre los dos posibles valores.

# Por otro lado el estado civil parece ser una variable significativa para predecir el estado de un empleado. Si analizamos los estados
# se puede ver que los estados civiles 2 y 4 tienen probabilidades similares de estar desvinculado, sin embargo estos estados civiles serían
# casado y soltero respectivamente, los cuales son los estados civiles mas comunes para la población, por lo que esta variable 
# no sería tan significativa para predecir el estado de un empleado, ya que su mayor frecuencia en la desvinculación se debe a que 
# son los estados civiles mas comunes y no a un factor a considerar en la desvinculación.

# En fin, se puede decir que el departamento y la posición son variables bastante significativas para predecir el estado de un empleado. 
```
