```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# leer archivo txt y crear dataframe
df = pd.read_csv('HousingData.csv')
df
```

```python
# Renombramos las columnas
df = df.rename(columns={'CRIM': 'Tasa_de_crimen'
                        , 'ZN': 'Proporción_de_terreno_residencial'
                        , 'INDUS': 'Proporción_de_acres_de_negocio'
                        , 'CHAS': 'Frontera_rio'
                        , 'NOX': 'Concentracion_oxido_nitroso'
                        , 'RM': 'Nro_habitaciones_promedio'
                        , 'AGE': 'Proporción_unidades_construidas_antes_1940'
                        , 'DIS': 'Distancia_promedio_centros_empleo'
                        , 'RAD': 'Indice_accesibilidad_autopistas'
                        , 'TAX': 'Tasa_impuesto_propiedad'
                        , 'PTRATIO': 'Tasa_alumnos_profesor'
                        , 'B': 'Proporción_negra'
                        , 'LSTAT': 'Porcentaje_clase_baja'
                        , 'MEDV': 'Valor_mediano_propiedad'})

df
```

# Histograma, Dispersión y Boxplot

```python
# Se crea una figura con 4 subplots
fig, axs = plt.subplots(2,2, figsize=(12,12))

# Se agrega el histograma a la figura
axs[0, 0].hist(df['Nro_habitaciones_promedio'])
axs[0, 0].set_title('Histograma de Nro habitaciones promedio')
axs[0, 0].set_xlabel('Nro habitaciones promedio')
axs[0, 0].set_ylabel('Frecuencia')

# La distribución del numero promedio de habitaciones tiene una campana leptocúrtica y simetrica, 
# ya que las mayores frecuencias se concentran mucho en el centro del gráfico empinando la campana
# y en los costados se observan menos frecuencias con colas mas similares.

# Se agrega grafico de dispersión
axs[0, 1].scatter(df['Valor_mediano_propiedad'], df['Tasa_de_crimen'])
axs[0, 1].set_title('Dispersión Valor mediano propiedad vs Tasa de crimen')
axs[0, 1].set_xlabel('Valor mediano propiedad')
axs[0, 1].set_ylabel('Tasa de crimen')

# Del gráfico se puede desprender a primera vista que la tasa de crimen tiende a ser mayor en las propiedades de menor valor, ya que los puntos
# mas elevados en el eje Y (mayor tasa de crimen) se encuentran más concentrados en la parte izquierda del gráfico. 
# Esto tiene cierta logica al considerar que en la realidad las propiedades de menor valor se encuentran en zonas con mayor delincuencia.

# Se agrega grafico de barras
# pendiente por falta de ciudades
axs[1, 0].bar(df['Frontera_rio'], df['Valor_mediano_propiedad'])
axs[1,0].set_title('Grafico Pendiente')


# Se agrega el boxplot
df['Tasa_de_crimen'].plot(kind='box',ax = axs[1,1],label='Tasa de crimen',showfliers=True) # showfliers=True muestra los outliers
axs[1, 1].set_title('Boxplot Tasa de crimen')

# Del boxplot se puede observar que existe una gran cantidad de outliers, ya que la caja se encuentra muy pequeña (entre 0 y 10) 
# y los bigotes son demasiado largos extendiendo el rango hasta valores cercanos a 90.

```
