```python
with open('videojuegos1.csv', 'r') as archivo:
    listado_lineas = []
    for linea in archivo.readlines():
        linea = linea.replace('"','')
        linea = linea.strip('\n')
        listado_lineas.append(linea)

    listado_registros = []
    for elemento in listado_lineas:
        registro = elemento.split(',')
        listado_registros.append(registro)

    nombre = []
    calificacion = []
    precio = []

    for valor in listado_registros:
        nombre.append(valor[0])
        calificacion.append(valor[1])
        precio.append(valor[2])

    diccionaro_df = {
        'Nombre': nombre[1:],
        'Calificacion': calificacion[1:],
        'Precio': precio[1:]
    }

videojuegos_df = pd.DataFrame(diccionaro_df)
videojuegos_df
```

```python
# Setear las variables del dataframe a float para evitar errores al calcular la media
videojuegos_df.Calificacion = videojuegos_df.Calificacion.astype(float)
videojuegos_df.Precio = videojuegos_df.Precio.astype(float)

media_precio = videojuegos_df['Precio'].mean()
media_calificacion = videojuegos_df['Calificacion'].mean()
print(f"El promedio de los precios es: {round(media_precio,2)}\nEl promedio de las calificaciones es: {round(media_calificacion,2)}")
```

```python
archivo2 = open('videojuegos2.csv', 'r')

lineas = archivo2.readlines()

listado_lineas = []
for linea in lineas:
    linea = linea.split('"')[1]
    linea = linea.replace('"','')
    linea = linea.replace('\n','')
    listado_lineas.append(linea)

listado_registros = []
for elemento in listado_lineas:
    registro = elemento.split(',')
    listado_registros.append(registro)

# agregar todas las columnas excepto tags que tiene una logica distinta al tener los registros separados por coma
nombre = []
year = []
consola = []
calificacion = []

for valor in listado_registros:
    nombre.append(valor[0])
    year.append(valor[1])
    consola.append(valor[2])
    calificacion.append(valor[3])


# logica para los tags
listado_tags = []
for linea in lineas:
    linea = linea.split('"')[-1] # se toma el ultimo elemento de la lista separada por comas para trabajar los tags
    linea = linea.strip(',')
    linea = linea.replace('\n','') 
    listado_tags.append(linea)
listado_tags[0] = 'Tag'
listado_tags

# logica para agregar la clasificación / categoria
columnas = []
for linea in lineas:
    columna = linea.split('"')[1] # se toma el primer elemento de la lista separada por comas para luego extraer el ultimo elemento categoria
    columnas.append(columna.split(','))

listado_clasificacion = []
for idx, columna in enumerate(columnas): # se recorre la columna para extraer el ultimo elemento y agregarlo al listado clasificacion categoria
    if idx == 0: # si la fila es 0 se omite ya que es el encabezado
        continue
    clasificacion = columna[-1] # este seria el último elemento de la columna que es deberia ir en clasificación categoria
    listado_clasificacion.append(clasificacion) # aqui se agrega el elemento al listado de calsificacion categoria

diccionaro_df = {
    'Nombre': nombre[1:],
    'Año': year[1:],
    'Consola': consola[1:],
    'Calificacion': calificacion[1:],
    'Tags': listado_tags[1:],
    'Categoria': listado_clasificacion
}

videojuegos2_df = pd.DataFrame(diccionaro_df)
videojuegos2_df.loc[0,'Calificacion'] = 3.5  # se corrije un dato extraño
videojuegos2_df_to_csv = videojuegos2_df.to_csv('videojuegos3.csv', index=False) # se guarda el df en un csv para poder leerlo en el siguiente ejercicio
```

```python
def juegosConsolas(nomArchivo, categoria, decada):
    df = pd.read_csv(nomArchivo)
    mascara_categoria = df['Categoria'] == categoria
    decada_inicio = ((decada // 10) * 10) % 100
    decada_fin = decada_inicio + 10
    mascara_decada = (df['Año']  >= decada_inicio) & (df['Año'] < decada_fin)
    df_filtro = df[mascara_categoria & mascara_decada]
    return df_filtro
```

```python
def crearMatriz(nomArchivo, categoria, decada):
    
    df = pd.read_csv(nomArchivo)
    mascara_categoria = df['Categoria'] == categoria
    decada_inicio = ((decada // 10) * 10) % 100
    decada_fin = decada_inicio + 10
    mascara_decada = (df['Año']  >= decada_inicio) & (df['Año'] < decada_fin)
    df_filtro = df[mascara_categoria & mascara_decada]
    df_filtro = df_filtro[['Nombre','Consola','Calificacion']]
    nueva_matriz = df_filtro.pivot_table(index='Nombre', columns='Consola', values='Calificacion')
    nueva_matriz= nueva_matriz.fillna(0) # para dejar los NaN como 0
    return nueva_matriz, df_filtro
```

```python
pivote,tabla = crearMatriz('videojuegos3.csv', 'RPG', 1986)
pivote2,tabla2 = crearMatriz('videojuegos3.csv', 'FPS', 1918)
tabla
```
