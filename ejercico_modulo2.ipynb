```python
# Volver a cargar los datafreames con separador ";"
df_abbrevs = pd.read_csv('state-abbrevs.csv', sep=";")
df_area = pd.read_csv('state-area.csv', sep=";")
df_population = pd.read_csv('state-population.csv', sep=";")
```

# 3.

```python
# Los datos de df_abbrevs y df_area solo tienen dos columnas con datos, por lo que se pueden seleccionar solo esas dos columnas
df_abbrevs = df_abbrevs.iloc[:,0:2]
df_area = df_area.iloc[:,0:2]
```

```python
df_abbrevs.head(8)
```

```python
df_area.head(8)
```

```python
# df_population tiene 4 columnas con datos, por lo que se seleccionan las 4 columnas
df_population = df_population.iloc[:,0:4]
df_population
```

# 4.

```python
# Agrupar los datos de df_population por state/region y ages y hacer agg por mean, min, max, count y sum
df_population_agrupado = df_population.groupby(['state/region','ages']).agg({'year':['mean','min','max','count','sum'],'population':['mean','min','max','count','sum']})
# renombrar mean como media, min como minimo, max como maximo y count como conteo
df_population_agrupado = df_population_agrupado.rename(columns={'mean':'Media','min':'Minimo','max':'Maximo','count':'Contar','sum':'Suma'})
df_population_agrupado.head(8)
```

# 5.

```python
# Hacer un merge de los dataframes df_abbrevs y df_area por state
df_abbrevs_area = pd.merge(df_abbrevs, df_area, on='state')
df_abbrevs_area.head(8)

# Hacer un merge de los dataframes df_abbrevs_area y df_population por state/region con abbreviation
df_abbrevs_area_population = pd.merge(df_abbrevs_area, df_population, left_on='abbreviation', right_on='state/region')
df_abbrevs_area_population.head(8)
```

```python
df_abbrevs_area_population.info()
df_abbrevs_area_population
```

```python
# Crear campo con population/area
df_abbrevs_area_population['Poblacion por km2'] = df_abbrevs_area_population['population'] / df_abbrevs_area_population['area (sq. mi)']
df_abbrevs_area_population
```

```python
# agruopar pot state y sum population
df_abbrevs_area_population_agrupado = df_abbrevs_area_population.groupby(['state']).agg({'population':['sum']})
df_abbrevs_area_population_agrupado
```

```python
# agrupar por state y Poblacion por km2 sin agg
df_abbrevs_area_population_agrupado = df_abbrevs_area_population[['state','Poblacion por km2']].groupby(['state'])['Poblacion por km2'].mean()
df_abbrevs_area_population_agrupado
```
