```python
# Volver a cargar los datafreames con separador ";"
df_abbrevs = pd.read_csv('state-abbrevs.csv', sep=";")
df_area = pd.read_csv('state-area.csv', sep=";")
df_population = pd.read_csv('state-population.csv', sep=";")
```

# 3.

```python
# Los datos de df_abbrevs y df_area solo tienen dos columnas con datos, por lo que se pueden seleccionar solo esas dos columnas
df_abbrevs = df_abbrevs.iloc[:,0:2]
df_area = df_area.iloc[:,0:2]
```

```python
df_abbrevs.head(8)
```

```python
df_area.head(8)
```

```python
# df_population tiene 4 columnas con datos, por lo que se seleccionan las 4 columnas
df_population = df_population.iloc[:,0:4]
df_population
```

# 4.

```python
# Agrupar los datos de df_population por state/region y ages y hacer agg por mean, min, max, count
df_population_agrupado = df_population.groupby(['state/region','ages']).agg({'population':['mean','min','max','count','sum']})
# renombrar mean como media, min como minimo, max como maximo y count como conteo
df_population_agrupado = df_population_agrupado.rename(columns={'mean':'Media','min':'Minimo','max':'Maximo','count':'Contar','sum':'Suma'})
df_population_agrupado.head(8)
```

# 5.

```python
# Hacer un merge de los dataframes df_abbrevs y df_area por state
df_abbrevs_area = pd.merge(df_abbrevs, df_area, on='state')

# Hacer un merge de los dataframes df_abbrevs_area y df_population por state/region con abbreviation
df_abbrevs_area_population = pd.merge(df_abbrevs_area, df_population, left_on='abbreviation', right_on='state/region')
df_abbrevs_area_population.head(8)
```

```python
df_abbrevs_area_population.info()
df_abbrevs_area_population
```

# 6.

```python
# Crear campo con population/area
df_abbrevs_area_population['Poblacion por area'] = df_abbrevs_area_population['population'] / df_abbrevs_area_population['area (sq. mi)']
df_abbrevs_area_population
```

```python
# agrupar por state y agg mean de Poblacion por area
df_abbrevs_area_population_agrupado = df_abbrevs_area_population[['state/region','state','Poblacion por area']].groupby(['state/region','state']).mean()
df_abbrevs_area_population_agrupado.head(8)
```

# 7.

```python
# cree un dataframe que posea: El estado, el área, la población, la población por área y como índice la abreviación del nombre del estado.
df_abbrevs_area_population_agrupado_2 = df_abbrevs_area_population[['abbreviation','state','area (sq. mi)','population','Poblacion por area']].groupby(['abbreviation','state']).mean()
df_abbrevs_area_population_agrupado_2.head(8)
```

```python
# se deben corregir los indices multiples para dejar solo la abreviacion como index
df_abbrevs_area_population_agrupado_2.index
```

```python
# separa los indices multiples en dos listas
lista_abreviaciones= []
lista_estados = []

for i in range(len(df_abbrevs_area_population_agrupado_2.index.values)):
    lista_abreviaciones.append(df_abbrevs_area_population_agrupado_2.index.values[i][0])
    lista_estados.append(df_abbrevs_area_population_agrupado_2.index.values[i][1])
```

```python
# se setea el indice y se agrega la columna state
df_abbrevs_area_population_agrupado_2.index = lista_abreviaciones
df_abbrevs_area_population_agrupado_2['state'] = lista_estados

# ordenar colummas
df_abbrevs_area_population_agrupado_3 = df_abbrevs_area_population_agrupado_2[['state','area (sq. mi)','population','Poblacion por area']]

# ordenar por poblacion por area
df_abbrevs_area_population_agrupado_3.sort_values(by=['Poblacion por area'], ascending=False)

# Se obteien que el estado con mayor poblacion por area es District of Columbia y el estado con menor poblacion por area es Alaska.
# Sin embargo, se debe destacar que el area de District of Columbia es de 68.34 millas cuadradas,
# mientras que el area de Alaska es de 656425.0 millas cuadradas.
```
