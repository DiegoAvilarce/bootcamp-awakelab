```python
import numpy as np
from scipy import stats
```

```python
consumos_empleados = [710, 732, 654, 760, 774, 675, 757, 718, 814, 755,736, 774, 768, 770, 677, 747, 823, 660, 797, 796]

n = 20
media = 721
```

```python
# hipotesis nula
valor_hipotesis = media

# one-sample t-test
_, p_value = stats.ttest_1samp(
    consumos_empleados, valor_hipotesis, alternative='greater' # greater para realizar test de cola derecha
)

# nivel de significancia del test de hipotesis
alpha = 0.1

# Evaluar si p_value es menor que alpha
if p_value < alpha:
    print("Hay evidencia suficiente para rechazar la hipotesis nula.")
else:
    print("No hay evidencia suficiente para rechazar la hipotesis nula, por lo que se asume como verdadera.")

# Hay evidencia suficiente para rechazar la hipotesis nula, por lo que se asume que el consumo promedio de los empleados es superior 
# a la media de consumo provincial de 721 kwh.

print("El p-value obtenido en este contraste de hipotesis es:", p_value)
```
