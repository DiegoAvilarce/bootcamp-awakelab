# 2. Separe sus datos en 70/30

```python
from sklearn.model_selection import train_test_split

# separar matriz x e y
X = datos[['tv', 'radio', 'periodico']]
y = datos.ventas

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)
```

# 3. Obtenga su modelo de Regresión lineal

```python
from sklearn.linear_model import LinearRegression
```

```python
# crear modelo lineal
modelo_lineal = LinearRegression()

# Ajustar el modelo
modelo_lineal.fit(X_train, y_train)
```

# 4. Entregue la recta matemática de regresión con las estimaciones obtenidas

```python
print(modelo_lineal.intercept_)
print(modelo_lineal.coef_)
print(f"La ecuación de la recta es: {modelo_lineal.intercept_:.3f} + {round(modelo_lineal.coef_[0],3)}*tv + {round(modelo_lineal.coef_[1],3)}*radio + {round(modelo_lineal.coef_[2],3)}*periodico")
```

# 5. Realice un diagnóstico de los residuos del modelo.

```python
prediccion_train = modelo_lineal.predict(X_train)
residuales_train = y_train - prediccion_train

# Graficar un histograma de los residuales
plt.hist(residuales_train)
```

```python
prediccion_test = modelo_lineal.predict(X_test)
residuales_test = y_test - prediccion_test

# Graficar un histograma de los residuales
plt.hist(residuales_test)
```

```python
prediccion = modelo_lineal.predict(X)
residuales = y- prediccion

# Graficar un histograma de los residuales
plt.hist(residuales)
```

# 6. Realice el test de normalidad

```python
# Test de hipotesis para verificar normalidad en los resiudales del modelo
from scipy.stats import shapiro
from scipy.stats import normaltest

# Shapiro-Wilk Test
_, p = shapiro(residuales_train)

print(f'Shapiro-Wilk Test: {p:.5f}')

if p > 0.05:
    print('Se rechaza H0, por lo que se puede asumir que los residuales del modelo distribuyen de forma normal.')

else:
    print('No se rechaza H0, por lo que los residuales del modelo no distribuyen de forma normal.')
```
