# F)

```python
# Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron
mascara_sobrevivientes = titanic_df['Survived'] == 1
mascara_no_sobrevivientes = titanic_df['Survived'] == 0
sobrevivientes = titanic_df[mascara_sobrevivientes]
nro_sobrevivientes = sobrevivientes['Survived'].count()
no_sobrevivientes = titanic_df[mascara_no_sobrevivientes]
nro_no_sobrevivientes = no_sobrevivientes['Survived'].count()
total_pasajeros = titanic_df['Survived'].count()
porcentaje_sobrevivientes = (nro_sobrevivientes / total_pasajeros) * 100
porcentaje_no_sobrevivientes = (nro_no_sobrevivientes / total_pasajeros) * 100

print(f"El porcentaje de sobrevicientes es: {round(porcentaje_sobrevivientes, 2)}%")
print(f"El porcentaje de no sobrevivientes es: {round(porcentaje_no_sobrevivientes, 2)}%")
```

# G)

```python
# Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase
for valor in titanic_df.Pclass.unique():
    mascara_sobrevivientes = titanic_df['Survived'] == 1
    mascara_clase = titanic_df['Pclass'] == valor
    total_pasajeros_clase = titanic_df[mascara_clase]['Pclass'].count()
    sobrevivientes_clase = titanic_df[mascara_sobrevivientes & mascara_clase]['Survived'].count()
    print(f"El porcentaje de sobrevivientes en la clase {valor} es: {round((sobrevivientes_clase / total_pasajeros_clase) * 100, 2)}%")
sobrevivientes_clase
```

# H)

```python
# Eliminar del DataFrame los pasajeros con edad desconocida
titanic_df.dropna(subset=['Age'], inplace=True) # inplace=True para que se modifique el dataframe original
```

# I)

```python
# Mostrar por pantalla la edad media de las mujeres que viajaban en cada clase
for valor in titanic_df.Pclass.unique():
    mascara_mujeres = titanic_df['Sex'] == 'female'
    mascara_clase = titanic_df['Pclass'] == valor
    mujeres_clase = titanic_df[mascara_mujeres & mascara_clase]
    edad_media_mujeres_clase = mujeres_clase['Age'].mean()
    print(f"La edad media de las mujeres en la clase {valor} es: {round(edad_media_mujeres_clase, 2)}")
```

# J)

```python
# Añadir una nueva columna booleana para ver si el pasajero era menor de edad o no
titanic_df['MenorEdad'] = titanic_df['Age'] < 18
titanic_df
```

# K)

```python
# Mostrar por pantalla el porcentaje de menores y mayores de edad que sobrevivieron en cada clase
for valor in titanic_df.Pclass.unique():
        mascara_menor_edad = titanic_df['MenorEdad'] == True
        mascara_mayor_edad = titanic_df['MenorEdad'] == False
        mascara_sobrevivientes = titanic_df['Survived'] == 1
        mascara_clase = titanic_df['Pclass'] == valor
        total_pasajeros_clase = titanic_df[mascara_clase]['Pclass'].count()
        pasajeros_menores = titanic_df[mascara_menor_edad & mascara_clase & mascara_sobrevivientes]['MenorEdad'].count()
        pasajeros_mayores = titanic_df[mascara_mayor_edad & mascara_clase & mascara_sobrevivientes]['MenorEdad'].count()
        print(f"El porcentaje de menores de edad que sobrevivieron en la clase {valor} es: {round((pasajeros_menores / total_pasajeros_clase) * 100, 2)}%")
        print(f"El porcentaje de mayores de edad que sobrevivieron en la clase {valor} es: {round((pasajeros_mayores / total_pasajeros_clase) * 100, 2)}%")
        
```
